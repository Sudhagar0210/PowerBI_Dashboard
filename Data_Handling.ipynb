Products 

```python
df_products = pd.read_csv("Products.csv",encoding='latin1')
df_products = pd.DataFrame(df_products)
df_products.head()
```

```python
df_products["Unit Cost USD"] = df_products["Unit Cost USD"].str.replace("$","").str.replace(",","")
df_products["Unit Cost USD"] = df_products["Unit Cost USD"].astype(float)

df_products["Unit Price USD"] = df_products["Unit Price USD"].str.replace("$","").str.replace(",","")
df_products["Unit Price USD"] = df_products["Unit Price USD"].astype(float)
```

```python
df_products.head()
```

Sales

```python
df_sales = pd.read_csv("Sales.csv",encoding='latin1')
df_sales = pd.DataFrame(df_sales)
df_sales.head()
```

```python
df_sales["Order Date"] = pd.to_datetime(df_sales["Order Date"],errors="coerce").dt.date
df_sales.head() 
```

```python
df_sales["Delivery Date"] = df_sales["Order Date"].fillna(df_sales["Order Date"])
df_sales.head()
```

```python
df_sales.isnull().sum()
```
